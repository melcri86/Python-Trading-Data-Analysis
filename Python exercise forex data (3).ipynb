```python
#Import pandas#
import pandas as pd
```

```python
#Import csv dataset#
data = pd.read_csv("D:\Courses\Dataset and Instructions\DATASET-A.csv")
data.head()
```

```python
#Import numpy#
import numpy as np
#replace NAN with blank space#
data1 = data.replace(np.nan, '', regex=True)
data1.head()
```

```python
#Replace wording BLANK with blank space#
data2 = data1.replace('BLANK', '', regex=True)
data2.head()
```

```python
#Count total number of rows of dataset#
len(data2)
```

```python
#Select desidered columns#
data3 = data2[['Date', 'Trader Id', 'OrderDirection','Security','OrderStatus','Price','Quantity']]
data3.head()
```

```python
#Rename column headers as desidered#
data4 = data3.rename(columns={"Trader Id": "Trader ID", "OrderDirection": "Order Direction", "OrderStatus":"Order Status"})
data4.head()
```

```python
#Sort dataset#
data5 = data4.sort_values(by=['Date','Trader ID','Order Direction','Security','Order Status','Price', 'Quantity'])
data5.head()
```

```python
#Filter dataset where Order Status equals TRADE#
data6 =data5.loc[(data5["Order Status"] == "TRADE")& (data5["Price"] != 0)]
data6.head()
```

```python
#Remove blank values from TRADER ID#
data7 = data6[data6['Trader ID'].values.astype(bool)]
data7.head()
```

```python
#Remove duplicated rows#
data8 = data7.drop_duplicates()
data8.head()
```

```python
#Sort new dataset based on certain columns#
data9 = data8.sort_values(by=['Date','Trader ID','Security','Price','Quantity'])
data9.reset_index()
```

```python
#Select matching rows by date, trader id, security, price and quantity#
data9['Match'] = data9.duplicated(subset=['Date','Trader ID','Security','Price','Quantity'], keep=False)
data9.head()
data10 = data9[data9['Match'] == True]
data10.head(50)
```

```python
#Format dataset#
from datetime import datetime
data10["Date"] = pd.to_datetime(data10["Date"], dayfirst=True).dt.strftime('%d-%b-%y')
data10["Quantity_MM"] = (data10['Quantity']/1000000).apply(lambda x: '{:,.2f}MM'.format(x))
data10["Price"]=data10["Price"].round(decimals=4)
data10["Security"]=data10["Security"].str.upper()
data10["Order Status"]=data10["Order Status"].str.title()
data10.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("white") 
sns.set_palette("pastel")

#Number of paired orders per day#
Count=data10['Date'].value_counts()
print('Number of paired orders per day')
print(Count)

Count1=data10['Date'].sort_values(ascending=True).value_counts().plot.bar(stacked=False, legend=None, title='Number of paired orders per day',sort_columns=True)
plt.savefig('Number of paired orders per day.png')
plt.show()
Count1b=data10['Date'].sort_values(ascending=True).value_counts().plot.line()
plt.savefig('Number of paired orders per day2.png')
plt.show()

#Number of transactions per order direction#
OrderDirection=data10['Order Direction'].value_counts()
print('Number of transactions per order direction')
print(OrderDirection)
OrderDirection1=data10['Order Direction'].value_counts().plot.bar(stacked=False, legend=None, title='Number of transactions per order direction',sort_columns=None)
plt.savefig('Number of transactions per order direction.png')
plt.show()

print('Transactions per day by order direction')
data_crosstab5 = pd.crosstab(data10['Date'],  
                             data10['Order Direction'], margins = False)
print(data_crosstab5) 
data_crosstab5a = pd.crosstab(data10['Date'],  
                             data10['Order Direction'], margins = False).plot.bar(stacked=False, legend=None, title='Transactions per day by order direction')
plt.savefig('Number of transactions per day by order direction.png')
plt.show()

#Unique dates#
dates = data10.Date.unique()
print('Unique Dates:')
print(dates)


#Unique traders#
traders = data10['Trader ID'].unique()
print('Unique trader ID')
print(traders)
#Number of unique traders#
Num_traders = data10['Trader ID'].nunique()
print('Number of unique trader ID')
print(Num_traders)
#Number of transactions per trader#
freq_traders = data10['Trader ID'].value_counts()
print('Number of paired orders per trader ID')
print(freq_traders)
freq_traders1 = data10['Trader ID'].value_counts().plot.bar(stacked=False, legend=None, title ='Number of orders per trader ID')
plt.savefig('Number of paired orders per trader ID.png')
plt.show()


#Unique security#
security = data10['Security'].unique()
print('Unique Security')
print(security)
#Number of unique security#
Num_Security = data10['Security'].nunique()
print('Number of unique Security')
print(Num_Security)
#Number of transactions per security#
freq_security = data10['Security'].value_counts()
print('Number of transactions per Security')
print(freq_security)
freq_security1 = data10['Security'].value_counts().plot.bar(stacked=False, legend=None, title='Number of orders per security')
plt.savefig('Number of transactions per security.png')
plt.show()

#Price statistics#
Avg_price = data10['Price'].mean()
print('The average price is:')
print(Avg_price)
Median_price = data10['Price'].median()
print('The median price is:')
print(Median_price)
Stdev_price = data10['Price'].std()
print('The standard deviation is:')
print(Stdev_price)
Max_price = data10['Price'].max()
print('The maximum price is:')
print(Max_price)
Min_price = data10['Price'].min()
print('The minimum price is:')
print(Min_price)

print('Distribution of Price')
data10.boxplot(column=['Price'])
plt.savefig('Distribution of Price.png')
plt.show()

# Grouped boxplot#
print('Distribution of Price per Day')
sns.boxplot(x="Date", y="Price", data=data10, palette="Set1")
plt.savefig('Distribution of Price per Day.png')
plt.show()

print('Distribution of Price per trader ID')
plt.figure(figsize=(20,10))
sns.boxplot(x="Trader ID", y="Price", data=data10, palette="Set1")
plt.savefig('Distribution of Price per trader ID.png')
plt.show()

print('Distribution of Price per security')
plt.figure(figsize=(20,5))
sns.boxplot(x="Security", y="Price", data=data10, palette="Set1")
plt.savefig('Distribution of Price per security.png')
plt.show()

print('Average price per day')
df_daily1=data10.groupby('Date').agg(['mean']).reset_index()
print(df_daily1)



#Quantity statistics#
Avg_q = data10['Quantity'].mean()
print('Average quantity')
print(Avg_q)
Median_q = data10['Quantity'].median()
print('Median quantity')
print(Median_q)
Stdev_q = data10['Quantity'].std()
print('Standard deviation quantity')
print(Stdev_q)
Max_q = data10['Quantity'].max()
print('Maximum quantity')
print(Max_q)
Min_q = data10['Quantity'].min()
print('Minimum quantity')
print(Min_q)
Sum_q = data10['Quantity'].sum()
print('Total quantity')
print(Sum_q)

print('Distribution of quantity')
data10.boxplot(column=['Quantity'])
plt.savefig('Distribution of quantity.png')
plt.show()

# Grouped boxplot#
sns.boxplot(x="Date", y="Quantity", data=data10, palette="Set1")
plt.savefig('Distribution of quantity per day.png')
plt.show()




print('Distribution of quantity per trader')

data_ct = pd.crosstab(data10['Trader ID'],  
                             data10['Quantity'], margins = False)
print(data_ct)
plt.figure(figsize=(20,10))
sns.boxplot(x="Trader ID", y="Quantity", data=data10, palette="Set1")
plt.savefig('quantity_per_trader.png')
plt.show()

print('Distribution of quantity per security')

data_ct = pd.crosstab(data10['Security'],  
                             data10['Quantity'], margins = False)
print(data_ct)
plt.figure(figsize=(20,10))
sns.boxplot(x="Security", y="Quantity", data=data10, palette="Set1")
plt.savefig('quantity_per_security.png')
plt.show()

print('Number of orders per security and trader ID - stacked')
data_crosstab = pd.crosstab(data10['Trader ID'],  
                             data10['Security'], margins = False)
print(data_crosstab)
data_crosstab11 = pd.crosstab(data10['Security'],  
                             data10['Trader ID'], margins = False).plot(kind='barh',stacked=True, legend=None, title='Number of orders per security and trader ID', sort_columns=True)
plt.savefig('Number of transaction per security and trader ID stacked.png')
plt.show()

#Number of transaction per security and trader ID#
print('Number of transaction per security and trader ID - not stacked')
dataa=data10[data10['Security'] == 'USD/INR']
data_crosstab11a = pd.crosstab(dataa['Security'],  
                             dataa['Trader ID'], margins = False).plot(kind='barh',stacked=False, title='Number of orders per security and trader ID - USD/INR', sort_columns=True)
plt.savefig('Number of transaction per security and trader ID not stacked.png')
plt.show()

print('Number of security orders per day')
data_crosstab1a= pd.crosstab(data10['Date'],  
                             data10['Security'], margins = False)

print(data_crosstab1a)

#Number of trasnsaciton per day per security#
print('Number of trasnsaciton per day per security')
data_crosstab1 = pd.crosstab(data10['Date'],  
                             data10['Security'], margins = False).plot.bar(stacked=False, legend=None, title='Number of trasnsaciton per day per security')
plt.savefig('Number of trasnsaciton per day per security.png')
plt.show()

#Number of transaction per day per quantity per trader ID#
print('Number of transaction per day per quantity per trader ID')
data_crosstab2 = pd.crosstab(data10['Date'],  
                             data10['Trader ID'], margins = False)
print(data_crosstab2) 
data_crosstab2a = pd.crosstab(data10['Date'],  
                             data10['Trader ID'], margins = False).plot(kind='line', subplots = True, stacked=False, figsize=(15,40), sharey=True, sharex=True, title='Number of transaction per day per quantity per trader ID')
plt.savefig('Number of transaction per day per quantity per trader ID.png')
plt.show()

#Transactions quantity by date#
print('Transactions quantity by date')
data_crosstab3 = pd.crosstab(data10['Quantity'],  
                             data10['Date'], margins = False)
print(data_crosstab3) 
data_crosstab3a = pd.crosstab(data10['Quantity'],  
                             data10['Date'], margins = False).plot.bar(stacked=False, figsize=(15,5), title='Orders quantity by date')
plt.savefig('Transactions quantity by date.png')
plt.show()

print('Orders quantity by trader ID')
data_crosstab7 = pd.crosstab(data10['Quantity'],  
                             data10['Trader ID'], margins = False)
print(data_crosstab7) 
data_crosstab7a = pd.crosstab(data10['Quantity'],  
                             data10['Trader ID'], margins = False).plot.bar(stacked=False, figsize=(15,5), title='Orders quantity per trader ID')
plt.savefig('Orders quantity by trader ID.png')
plt.show()

print('Orders quantity by security')
data_crosstab7 = pd.crosstab(data10['Quantity'],  
                             data10['Security'], margins = False)
print(data_crosstab7) 
data_crosstab7a = pd.crosstab(data10['Quantity'],  
                             data10['Security'], margins = False).plot.bar(stacked=False, figsize=(15,5), title='Orders quantity per security')
plt.savefig('Orders quantity by security.png')
plt.show()

#Transaction distribution per quantity, price, day and trader#
print('Transaction distribution per quantity, price, day and trader')
cmap = sns.cubehelix_palette(as_cmap=True)
g = sns.relplot(x="Date", y="Quantity", hue='Price', size='Price', col='Date' , row='Trader ID'
                 ,kind="scatter", data=data10)
plt.savefig('Transaction distribution per quantity, price, day and trader ID.png')


```
